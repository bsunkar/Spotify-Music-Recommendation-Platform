#### Create Track using POST API

```python
import pandas as pd
import requests
import json

# Read the CSV file
df = pd.read_csv('./../data/train.csv')

# Convert the DataFrame to a list of dictionaries
data = df.to_dict(orient='records')

# Define the API endpoint
url = 'http://localhost:8000/'

# Define the headers
headers = {
    'accept': 'application/json',
    'Content-Type': 'application/json'
}

# Loop through each record in the data
for record in data:
    # Make a POST request to the API endpoint
    response = requests.post(url, headers=headers, data=json.dumps(record))

    # Check the status of the request
    if response.status_code == 200:
        print(f"Successfully inserted record: {record['track_id']}")
    else:
        print("status_code : ",response.status_code)
        print("content : ",response.content)
        print(f"Failed to insert record: {record['track_id']}")
```

#### Update Track using PUT API

```python
import requests
import json
import sys

# Define the track_id
track_id = '5SuOikwiRyPMVoIQDJUgSV'

# Define the API endpoint
url = f'http://localhost:8000/{track_id}'

# Define the headers
headers = {
    'accept': 'application/json',
    'Content-Type': 'application/json'
}

try:
    # Make a GET request to the API endpoint
    response = requests.get(url, headers=headers)

    # Check the status of the GET request
    if response.status_code == 200:
        print(f"Successfully fetched record: {track_id}")
        # Convert the response content from bytes to dict
        data = json.loads(response.content)
        print("Old data : ", data)
    else:
        print("GET request failed")
        print("status_code : ",response.status_code)
        print("content : ",response.content)
        sys.exit()

    # Update the data
    for key in data.keys():
        old_value = data[key]
        new_value = input(f"Enter new value for {key} (old value: {old_value}, press enter to skip): ")
        if new_value:
            data[key] = new_value

    # Make a PUT request to the API endpoint
    response = requests.put(url, headers=headers, data=json.dumps(data))

    # Check the status of the PUT request
    if response.status_code == 200:
        print(f"Successfully updated record: {track_id}")
        # Convert the response content from bytes to dict
        response_dict = json.loads(response.content)
        print("Response content as dict : ", response_dict)      
    else:
        print("PUT request failed")
        print("status_code : ",response.status_code)
        print("content : ",response.content)
except Exception as e:
    print(f"An error occurred: {e}")
```
